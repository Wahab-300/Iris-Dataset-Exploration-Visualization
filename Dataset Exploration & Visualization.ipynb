# Task 1: Exploring and Visualizing the Iris Dataset

## Problem Statement
The goal of this task is to explore and visualize the Iris dataset to understand
its structure, distributions, and relationships between features using pandas,
matplotlib, and seaborn.

## Dataset
- Name: Iris Dataset
- Source: Seaborn built-in dataset
- Rows: 150 | Columns: 5
- Features: sepal_length, sepal_width, petal_length, petal_width, species

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from numpy import column_stack

# load data directly from sns
df = sns.load_dataset('iris')

# Basic Structure
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print(df.head())
```

```python
# Statistical summary of the dataset
print(df.describe())

# Check for missing values
print("\nMissing Values:\n", df.isnull().sum())
```

# Scatter Plots




```python
# Yes, I can see the histograms.
# Scatter plot — relationship between sepal length and petal length
plt.figure(figsize=(8,6))
sns.scatterplot(x='sepal_length', y='petal_length',hue='species', data=df, palette='bright')
plt.title('Scatter Plot: Sepal Length vs. Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot — relationship between sepal width and petal width
plt.figure(figsize=(8,6))
sns.scatterplot(x='sepal_width', y='petal_width',hue='species', data=df, palette='colorblind')
plt.title('Scatter Plot: Sepal Width vs. Petal Width')
plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```

# Histogram

```python
# Histogram — distribution of each feature
df.drop(columns='species').hist(bins=20, figsize=(10, 6), color='steelblue', edgecolor='black', grid=False)

plt.suptitle('Feature Distributions', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Box plot — spread and outliers for each feature by species
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, feature in enumerate(features):
    sns.boxplot(data=df, x='species', y=feature, palette='Set2', ax=axes[i//2][i%2])
    axes[i//2][i%2].set_title(f'{feature} by Species')

plt.suptitle('Box Plots — Feature vs Species', fontsize=14)
plt.tight_layout()
plt.show()
```

# HeatMap


```python
# Heatmap — correlation between all numeric features
sns.heatmap(df.drop(columns='species').corr(), annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()
```

# Conclusion

- The Iris dataset has no missing values and is clean.
- Petal length and petal width are highly correlated (0.96).
- Setosa species is clearly separable from Versicolor and Virginica.
- Petal features are more useful than sepal features for distinguishing species.
- Visualizations confirm that the dataset is well-structured for classification tasks.
